## Ordenamos por tiempo y reemplazmos nan's por media

```python
df = pd.concat([df.timestamp,df.groupby('timestamp').transform(lambda x: x.fillna(x.mean()))],axis=1) #nope
```

```python
df.head()
```

** se eliminan columnas que quedan sin datos en algún timestamp**

```python
suma = 0
for time in df.timestamp.unique(): #con groupby
    sample =  df[df.timestamp==time]
    for i in sample.columns:
        if sum(sample[i].isnull()==True):
            suma+=1
            print "columna con solo null eliminada",i
            df = df.drop(i, axis=1)
print "total eliminados ",suma
```
